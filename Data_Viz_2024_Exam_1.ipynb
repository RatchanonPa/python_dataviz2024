```python
# prompt: drop column ' @dropdown'

water_df = water_df.drop(columns=['@dropdown'])
water_df
```

```python
# prompt: แก้ค่าใน column 'month' โดยแก้จาก July เป็น Jul และ June เป็น Jun
water_df['month'] = water_df['month'].replace('July_2018', 'Jul_2018')
water_df['month'] = water_df['month'].replace('July_2019', 'Jul_2019')
water_df['month'] = water_df['month'].replace('June_2019', 'Jun_2019')
water_df
```

```python
# Convert 'month' to datetime for easier manipulation
water_df['month'] = pd.to_datetime(water_df['month'], format='%b_%Y')

# Extract year from the 'month' column
water_df['year'] = water_df['month'].dt.year
water_df
```

```python
# prompt: ตัดวันที่ออกให้เหลือแค่เดือนและปี

# Format the 'month' column to show only month and year
water_df['month'] = water_df['month'].dt.strftime('%b_%Y')
water_df
```

# 1.หาความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
# prompt: หาความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

# Calculate mean Chlo_a for each station in 2018
chlo_a_2018 = water_df[water_df['year'] == 2018].groupby('station')['Chlo_a'].mean()

# Calculate mean Chlo_a for each station in 2019
chlo_a_2019 = water_df[water_df['year'] == 2019].groupby('station')['Chlo_a'].mean()

# Calculate the difference
difference = chlo_a_2019 - chlo_a_2018

difference
```

```python
# prompt: ทำให้เป็น dataframe โดยประกอบด้วย 3 คอลัมน์ คือ chlo_a_2018, chlo_a_2019, and difference.

import pandas as pd
# Create a DataFrame from the calculated values
difference_df = pd.DataFrame({
    'chlo_a_2018': chlo_a_2018,
    'chlo_a_2019': chlo_a_2019,
    'difference': difference
})

difference_df
```

# 2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
# prompt: สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน แบ่ง station เป็น 6 columns

# สร้างตาราง Pivot Table
pivot_table = water_df.pivot_table(index='month', columns='station', values='Chlo_b', aggfunc='mean')

# แสดงผลลัพธ์
pivot_table
```

```python
# prompt: เรียง pivot table โดยเริ่มจาก Jul_2018 และสิ้นสุดที่ Jul_2019

# Sort the pivot table by month
pivot_table = pivot_table.reindex(['Jul_2018',
                                   'Aug_2018',
                                   'Sep_2018',
                                   'Oct_2018',
                                   'Nov_2018',
                                   'Dec_2018',
                                   'Jan_2019',
                                   'Apr_2019',
                                   'May_2019',
                                   'Jun_2019',
                                   'Jul_2019'])

# แสดงผลลัพธ์
pivot_table
```
