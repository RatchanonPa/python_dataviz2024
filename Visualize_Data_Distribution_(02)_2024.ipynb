# การแสดงผลการกระจายของข้อมูล

````
import pandas as pd
from matplotlib import pyplot as plt
````

````
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df.head()
````

## Scatter 2D

## scatter - default

````
plt.scatter(df['PetalLength'],df['PetalWidth'])
````

### color

````
plt.scatter(df['PetalLength'],df['PetalWidth'],c='r')
````

## ใส่สีให้แต่ละจุด

````
set(df['Name']) #ดูว่ามีค่าอะไรบ้าง
````

````
df2 = df.replace({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica':2})
df2
````

````
plt.scatter(df['PetalLength'],df['PetalWidth'],c=df2['Name'])
````

### alpha กำหนดความโปร่งแสง
เวลา plot scatter กำหนด alpha ด้วยเนื่องจากอาจมีบางจุดที่ทับกันอยู่อาจทำให้เราเข้าใจ data ผิดได้

````
plt.scatter(df['PetalLength'],df['PetalWidth'],c=df2['Name'],alpha=0.5)
````

## ใช้ขนาดของ marker แสดงค่าของ feature

````
plt.scatter(df['PetalLength'],df['SepalLength'],s=50*df['PetalWidth'],c=df2['Name'],alpha=0.5)
````

````
plt.scatter(df['PetalLength'],df['PetalWidth'],s=10*df['SepalWidth'],c=df2['Name'],alpha=0.5)
````

## ตัวแปร Marker ใช้ในการเปลี่ยนหน้าตาของ marker

````
plt.scatter(df['PetalLength'],df['PetalWidth'],s=100,c=df2['Name'],alpha=0.5,marker='*')
````

## การ plot กราฟหลายๆอันซ้อนกัน

ถ้าไม่ set alpha กราฟแรกที่ plot จะโดนทับ

````
plt.scatter(df['PetalLength'][:50],df['PetalWidth'][:50],s=100,c='m',alpha=0.5,marker='$S$') # s = Size
plt.scatter(df['PetalLength'][50:100],df['PetalWidth'][50:100],s=100,c='c',alpha=0.5,marker='$E$')
plt.scatter(df['PetalLength'][100:],df['PetalWidth'][100:],s=100,c='r',alpha=0.5,marker='$I$')
````

###Label-Legend

```plt.plot()``` input  ตัวที่ 3 คือ ตัวกำหนดหน้าตาของ marker

'หน้าตาmarker'-'ลักษณะของเส้นเชื่อมจุด'-'สี'

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5)
#x, y,'or' is a format string that specifies: 'o' for circular markers,'r' for red color
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'o-r',alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'o:r',alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5,label='setosa')
plt.plot(df['PetalLength'][50:100],df['PetalWidth'][50:100],'xc',alpha=0.5,label='versicolor')
plt.plot(df['PetalLength'][100:],df['PetalWidth'][100:],'*m',alpha=0.5,label='virginica')
plt.legend(loc='lower right')#plt.legend() หา location ที่ดีที่สุด
````

### การใส่ชื่อกราฟ และชื่อแกน

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5,label='setosa')
plt.plot(df['PetalLength'][50:100],df['PetalWidth'][50:100],'xc',alpha=0.5,label='versicolor')
plt.plot(df['PetalLength'][100:],df['PetalWidth'][100:],'*m',alpha=0.5,label='virginica')
plt.legend(loc='lower right')
plt.title('Iris flower data distribution') ##ชื่อกราฟ
plt.xlabel('PetalLength') ##ชื่อแกน X
plt.ylabel('PetalWidth'); ##ชื่อแกน y
````

## Data Preprocessing

````
import pandas as pd
import os
import matplotlib.pyplot as plt
````

````
from google.colab import drive
drive.mount('/content/drive')
````

````
path = '/content/drive/MyDrive/data_viz_2024_DATA'
````

````
movie_file_path = os.path.join(path,'IMDb Movies India.csv')
print(movie_file_path)
````

````
movie_df = pd.read_csv(movie_file_path, encoding='latin-1') # try a different encoding such as latin-1
movie_df
````

````
movie_df = movie_df.dropna()
movie_df = movie_df.drop_duplicates()
print(movie_df.info())
movie_df
````

````
# Delete min character from 'Duration' column on movie_df
movie_df['Duration'] = movie_df['Duration'].str.replace('min','')
# prompt: delete parentheses in 'Year' column on movie_df
movie_df['Year'] = movie_df['Year'].str.replace('(', '').str.replace(')', '')
print(movie_df.info())
movie_df
````

````
# Convert 'Votes' column to string type before using .str accessor
movie_df['Votes'] = movie_df['Votes'].astype(str).str.replace(',', '').astype(float)
# prompt: Convert 'Duration' column to float type
movie_df['Duration'] = movie_df['Duration'].astype(float)
# prompt: convert 'Year' column in movie_df to date type
movie_df['Year'] = pd.to_datetime(movie_df['Year'], format='%Y')
print(movie_df.info())
print(movie_df.describe())
movie_df
````

## Box plot
- Boxplot Rating of Indian Movies

````
movie_df
````

````
# ขยายขนาดกราฟ
plt.figure(figsize=(12, 6))  # กำหนดขนาดของกราฟ (กว้าง, สูง)

# สร้าง boxplot
O = plt.boxplot(movie_df['Rating'], showmeans=True, meanline=True, vert=False)

# เพิ่มชื่อแกน
plt.xlabel('Rating')
plt.title('Distribution of IMDb Ratings for Indian Movies')

# แสดงกราฟ
plt.show()
````

## Top 3 Genre Movies

- เรียงลำดับจำนวนประเภทหนังแต่ละประเภท

````
df_genre = pd.DataFrame(movie_df, columns=['Genre'])
df_genre = df_genre.groupby('Genre').size() # Use size() to count occurrences in each group
df_genre = df_genre.reset_index(name='Count')
df_genre = df_genre.sort_values(by='Count', ascending=False)
df_genre
````

- สร้าง dataframe ใหม่ที่เก็บ top 3 genres

````
# Filter the DataFrame to include rows where the 'Genre' column is one of the desired values
movie_genre_top3_df = movie_df[movie_df['Genre'].isin(['Drama', 'Drama, Romance', 'Action, Crime, Drama'])]
movie_genre_top3_df = movie_genre_top3_df[['Genre', 'Duration','Rating', 'Votes']] # Pass column names as strings in a list. Changed 'Rating' to 'Ratings'.
print(movie_genre_top3_df.info())
movie_genre_top3_df
````

````
# prompt: group by Genre in movie_genre_top3_df
movie_genre_top3_df.groupby('Genre').describe()
````

- สร้าง dataframe ใหม่ที่เก็บ top 3 genres โดยสุ่มข้อมูลมา 300 จุด

````
movie_genre_top3_rand_df = movie_genre_top3_df.sample(300) # Use sample() to get a random row from the DataFrame
print(movie_genre_top3_rand_df.info())
movie_genre_top3_rand_df
````

````
movie_genre_top3_rand_df.groupby('Genre').describe()
````

## Scatter plot


- Top 3 Indian IMDB Genre data distribution between Rating and Duration

````
# Scatter plot for each genre
plt.scatter(movie_genre_top3_df[movie_genre_top3_df['Genre'] == 'Drama']['Rating'],
            movie_genre_top3_df[movie_genre_top3_df['Genre'] == 'Drama']['Duration'],
            color='red', marker='o', alpha=0.4, label='Drama')

plt.scatter(movie_genre_top3_df[movie_genre_top3_df['Genre'] == 'Drama, Romance']['Rating'],
            movie_genre_top3_df[movie_genre_top3_df['Genre'] == 'Drama, Romance']['Duration'],
            color='cyan', marker='x', alpha=0.4, label='Drama, Romance')

plt.scatter(movie_genre_top3_df[movie_genre_top3_df['Genre'] == 'Action, Crime, Drama']['Rating'],
            movie_genre_top3_df[movie_genre_top3_df['Genre'] == 'Action, Crime, Drama']['Duration'],
            color='magenta', marker='*', alpha=0.4, label='Action, Crime, Drama')

# Add labels and title
plt.legend()
plt.title('Top 3 Indian IMDb Genres Data Distribution between Rating and Duration')
plt.xlabel('Rating')
plt.ylabel('Duration')

# Show plot
plt.show()
````
