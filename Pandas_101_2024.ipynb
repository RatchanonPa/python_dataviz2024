### Process Overview

1. **Load the Data**:
   - Read the dataset from the provided CSV file.
   
2. **Clean the Data**:
   - Remove HTML tags from the `description` column.
   - Unescape HTML entities.
   - Remove special symbols from the cleaned descriptions.

3. **Calculate Description Lengths**:
   - Compute the length of each cleaned description to identify the longest and shortest stories.

4. **Analyze Story Types**:
   - Count the occurrences of each unique value in the `type` column to determine the most and least common story types.

5. **Count Non-Thai Spatial Stories**:
   - Filter the dataset to count the number of stories where the `spatial` value is not 'ไทย' (Thai).

### Detailed Steps with Code

1. **Load the Data**:
    ```python
    import pandas as pd

    # Load the dataset
    file_path = '/mnt/data/folktales.csv'
    folk_df = pd.read_csv(file_path)
    ```

2. **Clean the Data**:
    ```python
    from html import unescape

    # Fill NaN values in the description column with empty strings
    folk_df['description_clean'] = folk_df['description'].fillna('')

    # Remove HTML tags
    folk_df['description_clean'] = folk_df['description_clean'].str.replace(r'<[^>]*>', '', regex=True)

    # Unescape HTML entities
    folk_df['description_clean'] = folk_df['description_clean'].apply(unescape)

    # Remove special symbols
    folk_df['description_clean'] = folk_df['description_clean'].str.replace(r'[^\w\s]', '', regex=True)
    ```

3. **Calculate Description Lengths**:
    ```python
    # Calculate the length of each description
    folk_df['description_length'] = folk_df['description_clean'].str.len()
    ```

4. **Identify Longest and Shortest Stories**:
    ```python
    # Identify the longest and shortest stories
    longest_story = folk_df.loc[folk_df['description_length'].idxmax()]
    shortest_story = folk_df.loc[folk_df['description_length'].idxmin()]
    ```

5. **Analyze Story Types**:
    ```python
    # Determine the most and least common types of stories
    story_type_counts = folk_df['type'].value_counts()
    most_common_type = story_type_counts.idxmax()
    least_common_type = story_type_counts.idxmin()
    ```

6. **Count Non-Thai Spatial Stories**:
    ```python
    # Count the number of spatial stories that are not Thai
    non_thai_spatial_stories_count = folk_df[folk_df['spatial'] != 'ไทย'].shape[0]
    ```

````
import pandas as pd
from html import unescape
````

````
# 1. นิทานที่ยาวที่สุด และสั้นที่สุด(ไม่นับสัญลักษณ์พิเศษ)
# Clean the 'description' column by removing HTML tags and special symbols
folk_df['description_clean'] = folk_df['description'].fillna('')  # Fill NaN values with empty strings
folk_df['description_clean'] = folk_df['description_clean'].str.replace(r'<[^>]*>', '', regex=True)
folk_df['description_clean'] = folk_df['description_clean'].apply(unescape)  # Unescape HTML entities
folk_df['description_clean'] = folk_df['description_clean'].str.replace(r'[^\w\s]', '', regex=True)
````

````
folk_df['description_clean'].head(5)
````

````
# Calculate the length of each description
folk_df['description_length'] = folk_df['description_clean'].str.len()
````

````
# Identify the longest and shortest stories
longest_story = folk_df.loc[folk_df['description_length'].idxmax()]
shortest_story = folk_df.loc[folk_df['description_length'].idxmin()]
````

````
# Output the results
print(f"นิทานที่ยาวที่สุดคือ {longest_story['title']} โดยมีความยาว {longest_story['description_length']} ตัวอักษร")
print("-------------------------------------------------------------------------------------------------------------------")
print(f"นิทานที่สั้นที่สุดคือ {shortest_story['title']} โดยมีความยาว {shortest_story['description_length']} ตัวอักษร")
````
