## Quize9

*   วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่องทั้งหมดกี่เรื่อง และเป็นเรื่องอะไรบ้าง
*   มีคำดังกล่าวใน นิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
* คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่หรือพ่อมากที่สุด


````
#prompt: วนลูปนับว่ามีนิทานที่มีคำว่า แม่หรือพ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

count_mom_dad = 0
list_story = []
for i, each_row in this_data.iterrows():
  if 'แม่' in each_row['title'] or 'พ่อ' in each_row['title']:
    count_mom_dad += 1
    list_story.append(each_row['title'])
print(f'มีนิทานที่มีคำว่า แม่หรือพ่อ ในชื่อเรื่อง ทั้งหมด {count_mom_dad} เรื่อง')
print(f'รายชื่อนิทาน: {list_story}')
````

````
# prompt: มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานชีวิตหรือนิทานประจำถิ่นมากกว่ากัน

count_life = 0
count_local = 0
for i, each_row in this_data.iterrows():
  if 'แม่' in each_row['title'] or 'พ่อ' in each_row['title']:
    if each_row['type'] == 'นิทานชีวิต':
      count_life += 1
    elif each_row['type'] == 'นิทานประจำถิ่น':
      count_local += 1

if count_life > count_local:
  print('มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานชีวิตมากกว่า')
elif count_local > count_life:
  print('มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานประจำถิ่นมากกว่า')
else:
  print('มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานชีวิตและนิทานประจำถิ่นเท่ากัน')
````

````
# prompt: คนเเต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่หรือพ่อ มากที่สุด

creator_counts = {}
for i, each_row in this_data.iterrows():
  if 'แม่' in each_row['title'] or 'พ่อ' in each_row['title']:
    creator = each_row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

max_count = 0
most_frequent_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่หรือพ่อ มากที่สุดคือ: {most_frequent_creator} ({max_count} เรื่อง)")
````
